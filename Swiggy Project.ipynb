4. Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" to calculate TOTAL ORDERS for each RESTAURANT_ID in Cuisine = Juices 

```python
pd.read_sql('''SELECT   RESTAURANT_ID as Juice_Restaurents_ID, TOTAL_ORDERS
FROM(
SELECT O.RESTAURANT_ID, R.CUISINE, COUNT(O.ORDER_ID) as TOTAL_ORDERS
FROM Order_Fact O
LEFT JOIN  Restaurant_Attributes R
ON O.RESTAURANT_ID=R.RESTAURANT_ID
GROUP BY O.RESTAURANT_ID, R.CUISINE
) A
WHERE CUISINE in  ('Juices')''',myconn)
```

5. Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" to calculate TOTAL REVENUE for each RESTAURANT_ID in Cuisine = Biryani

```python
pd.read_sql('''SELECT  RESTAURANT_ID as Restaurents_Biryani, TOTAL_REVENUE
FROM(
SELECT O.RESTAURANT_ID, R.CUISINE, SUM(O.ORDER_VALUE) as TOTAL_REVENUE
FROM Order_Fact O
left join Restaurant_Attributes R
on O.RESTAURANT_ID=R.RESTAURANT_ID
GROUP BY O.RESTAURANT_ID,R.CUISINE
) A
WHERE CUISINE in  ('Biryani')''',myconn)
```

6. Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" to calculate TOTAL ORDERS for each RESTAURANT_ID in Area = Koramangala 

```python
pd.read_sql('''SELECT  RESTAURANT_ID as Koramangala_Restaurents, TOTAL_ORDERS
FROM(
SELECT O.RESTAURANT_ID, R.AREA, COUNT(O.ORDER_ID) as TOTAL_ORDERS
FROM Order_Fact O
left join Restaurant_Attributes R
on O.RESTAURANT_ID=R.RESTAURANT_ID
GROUP BY O.RESTAURANT_ID, R.AREA
) A
WHERE AREA in  ('Koramangala')''',myconn)
```

7. Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" to calculate TOTAL ORDERS and AVERAGE RATING for each RESTAURANT_ID in Area = Koramangala and Restaurant_Category = A

```python
pd.read_sql('''SELECT   RESTAURANT_ID as Koramangala_A_RestaurentsID, TOTAL_ORDERS, Avg_Rating
FROM(
 SELECT O.RESTAURANT_ID, R.AREA, R.RESTAURANT_CATEGORY,
 COUNT(O.ORDER_ID) as TOTAL_ORDERS, ROUND(AVG(Rating),2) as Avg_Rating
 FROM Order_Fact O
 LEFT JOIN  Restaurant_Attributes R
 ON O.RESTAURANT_ID=R.RESTAURANT_ID
 GROUP BY O.RESTAURANT_ID, R.AREA,R.RESTAURANT_CATEGORY
) A
WHERE AREA in  ('Koramangala') and RESTAURANT_CATEGORY='A'
Order by Avg_Rating desc
''',myconn)
```

8. Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" to calculate TOTAL ORDERS and AVERAGE RATING for each RESTAURANT_ID in Area = Indiranagar and Restaurant_Category = B and C

```python
pd.read_sql('''SELECT   RESTAURANT_ID as Indiranagar_RestaurentsID, TOTAL_ORDERS, Avg_Rating
FROM(
 SELECT O.RESTAURANT_ID, R.AREA, R.RESTAURANT_CATEGORY,
 COUNT(O.ORDER_ID) as TOTAL_ORDERS, ROUND(AVG(Rating),2) as Avg_Rating
 FROM Order_Fact O
 LEFT JOIN  Restaurant_Attributes R
 ON O.RESTAURANT_ID=R.RESTAURANT_ID
 GROUP BY O.RESTAURANT_ID, R.AREA,R.RESTAURANT_CATEGORY
) A
WHERE AREA ='Indiranagar' and RESTAURANT_CATEGORY in ('B','C')
Order by Avg_Rating desc''',myconn)
```

9. Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" to calculate Total number of restaurants in an Area = HSR with Rating >4.0 and AOV >200 and Total Orders >150

```python
pd.read_sql('''SELECT Count(R.RESTAURANT_ID) as Total_Restaurents_in_HSR
FROM(
 SELECT RESTAURANT_ID,
 COUNT(ORDER_ID) as TOTAL_ORDERS
 FROM Order_Fact 
 GROUP BY RESTAURANT_ID) A
RIGHT JOIN Restaurant_Attributes R
ON A.RESTAURANT_ID=R.RESTAURANT_ID
WHERE AREA= 'HSR' AND Rating > 4 AND AOV > 200 AND TOTAL_ORDERS > 150''',myconn)
```

```python
#Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" to 
#calculate TOTAL ORDERS for each RESTAURANT_ID
both.groupby('RESTAURANT_ID').ORDER_ID.count()
```

```python
#Write the SQL query using the tables "Restaurant_Attributes" and "Order_Fact" 
#to calculate TOTAL REVENUE for each RESTAURANT_ID
both.groupby('RESTAURANT_ID').ORDER_VALUE.sum()
```

```python
#calculate TOTAL DISTINCT CUSTOMERS for each RESTAURANT_ID

both.groupby("RESTAURANT_ID").agg({"CUSTOMER_ID": np.sum, "CUSTOMER_ID": pd.Series.nunique})
```

```python
#calculate TOTAL DISTINCT CUSTOMERS for each RESTAURANT_ID

both.groupby("RESTAURANT_ID").agg({"CUSTOMER_ID": np.sum, "CUSTOMER_ID": lambda x: x.nunique()})
```

```python
#calculate TOTAL REVENUE for each RESTAURANT_ID in Cuisine = Biryani
both[both.CUISINE=='Biryani'].groupby('RESTAURANT_ID').ORDER_VALUE.sum()
```

```python
# calculate TOTAL ORDERS for each RESTAURANT_ID in Area = Koramangala 
both[both.AREA=='Koramangala'].groupby('RESTAURANT_ID').ORDER_ID.count()
```

```python
#calculate TOTAL ORDERS and AVERAGE RATING for each RESTAURANT_ID in Area = Koramangala 
#and Restaurant_Category = B
both[(both.AREA=='Koramangala' )&
     (both.RESTAURANT_CATEGORY=='B')].groupby('RESTAURANT_ID').ORDER_ID.count()
#.agg({'RESTAURANT_ID':np.sum, "RESTAURANT_ID": lambda x: x.nunique()})
```

```python
both[(both.AREA=='Koramangala' )&
     (both.RESTAURANT_CATEGORY=='B')].groupby('RESTAURANT_ID').agg(
    {'ORDER_ID':'count',
    'Rating':'mean'})
```

```python
#calculate TOTAL ORDERS and AVERAGE RATING for each RESTAURANT_ID in Area = Indiranagar
#and Restaurant_Category = B and C
both[(both.AREA=='Indiranagar' )&
     ((both.RESTAURANT_CATEGORY=='B') | (both.RESTAURANT_CATEGORY=='C'))].groupby('RESTAURANT_ID').agg(
    {'ORDER_ID':'count',
    'Rating':'mean'})
```
